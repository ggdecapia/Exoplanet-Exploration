```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("Resources\exoplanet_data.csv")
df
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
df
```

```python
# Drop the null rows
df = df.dropna()
df.columns
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
X = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 'koi_time0bk', 'koi_impact', 'koi_duration',
       'koi_depth', 'koi_prad', 'koi_teq', 'koi_insol', 'koi_model_snr', 
       'koi_tce_plnt_num', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec', 'koi_kepmag']]
X
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
y = df[['koi_disposition']]
y
```

```python
# y = pd.get_dummies(y)
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
print(X.shape, y.shape)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
# from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

label_encoder = LabelEncoder()
label_encoder.fit(y_train)
y_train_encoded = label_encoder.transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear')
```

```python
model.fit(X_train_scaled, y_train_encoded)
```

# Train the Model



```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model

# Create the SVC Model
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model
```

```python
model.fit(X_train_scaled, y_train_encoded)
```

```python
## Create the GridSearch estimator along with a parameter object containing the values to adjust
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10, 50],
              'gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch

grid.fit(X_train_scaled, y_train_encoded)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
